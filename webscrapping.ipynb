```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the IMDb list
url = 'https://www.imdb.com/list/ls072933289/'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the container that holds the list of movies
    movie_container = soup.find('div', class_='lister-list')

    # Find all the list items (each representing a movie)
    movie_items = movie_container.find_all('div', class_='lister-item')

    # Iterate through the movie items and extract information
    for movie in movie_items:
        # Extract the movie title
        title = movie.find('h3', class_='lister-item-header').a.text.strip()

        # Extract the IMDb rating
        rating = movie.find('span', class_='ipl-rating-star__rating').text.strip()

        # Print the movie title and rating
        print(f"Title: {title}, Rating: {rating}")

else:
    print('Failed to fetch the webpage.')
```
